```python
import pandas as pd
import statsmodels.formula.api as smf


model = smf.ols('price ~ area', data = data)
results = model.fit()
print(results.summary())
"""The OLS regression results indicate that the model explains approximately 28.7% of the variance in property prices (R-squared = 0.287). The coefficient for the area is 461.9749, meaning that for each additional unit of area, the price increases by approximately 462 units, holding other factors constant. The intercept is 2.387e+06, representing the baseline price when the area is zero. The model is statistically significant with a p-value of 7.39e-42."""
```

# The OLS regression results show that the model explains about 28.7% of the variance in property prices, which is a moderate level of explanatory power. The coefficient for the area is significant, indicating that the area has a strong positive impact on the price. However, the R-squared value suggests that there are other factors influencing property prices that are not included in this model.

```python
def Predict_Price(area, bedrooms, bathrooms, parking, basement, mainroad): # pridiction with multipal variables
    intercept = -6.599e+05
    coef_area = 298.1750
    coef_bedrooms = 3.92e+05
    coef_bathrooms = 1.304e+06
    coef_parking = 3.092e+05
    coef_basement = 4.148e+05
    coef_mainroad = 8.058e+05
    
    return (intercept + coef_area * area + coef_bedrooms * bedrooms + coef_bathrooms * bathrooms + coef_parking * parking + coef_basement * basement + coef_mainroad * mainroad)

# Example usage
print(Predict_Price(1650, 2, 1, 2, 1, 1))
print(Predict_Price(2400, 3, 1, 0, 0, 0))
print(Predict_Price(3620, 2, 1, 0, 1, 1))
```
